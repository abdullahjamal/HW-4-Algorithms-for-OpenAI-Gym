```python
import gym
import tensorflow as tf
import numpy as np
```

```python
env = gym.make('CartPole-v1')

def policy(state): #if angle is greater than zero, we move the pole to the right
    if state[2]  >  0:
        return  1  
    else:
        return 0
```

```python
max_episodes = 1000
max_t = 200
push_left = 0
push_right = 1
episode_reward_list = []
for i in np.arange(max_episodes):
    
    state = env.reset()
    env.render()
    total_reward_test = 0.0
    action  = 1#policy(state)
    for s in np.arange(max_t):
        state,reward,done,_ = env.step(action)
        action  = policy(state)
        
        total_reward_test += reward
        env.render()
        if done:
            break

    env.close()
    print("Total reward:",total_reward_test)
    episode_reward_list.append(total_reward_test)
```

```python
def running_mean(x, N):
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / N
```

```python
print("Average Reward:",sum(episode_reward_list)/len(episode_reward_list))
print("Max Reward:",max(episode_reward_list))
print("Min Reward:",min(episode_reward_list))
import matplotlib.pyplot as plt
rews = np.array(episode_reward_list).T
smoothed_rews = running_mean(rews, 10)
plt.plot(smoothed_rews)
plt.plot(rews, color='grey', alpha = 0.3)
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.show()
```
